```python
from article_scraper import QiuShiArticleParser, QiuShiArticle
```

```python
article_parser = QiuShiArticleParser("http://www.qstheory.cn/dukan/qs/2023-10/31/c_1129947183.htm")

article = QiuShiArticle(article_parser)
```

```python
text = article.text
punc_free_text = article.text
```

```python
chinese_punc = ''' ，。：；‘“’”！？（） 【】「」/、｜'''

# get rid of punctuation
for ele in punc_free_text:
    if ele in chinese_punc:
        punc_free_text = punc_free_text.replace(ele, "")
```

```python
# parse the Chinese
import jieba

seg_text = jieba.cut(punc_free_text)
seg_list = [word for word in seg_text]
```

```python
# create a dictionary with word counts

word_count = {}

for word in seg_list:
    if word in word_count.keys():
        cur_val = word_count[word]
        cur_val += 1
        word_count[word] = cur_val
    else:
        word_count[word] = 1
```

```python
word_count = {k: v for k, v in sorted(word_count.items(), key=lambda item: item[1], reverse=True)}
```
