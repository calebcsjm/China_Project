```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append('..')
from articles.analyze_articles.analysis_helper_functions import *
```

```python
qiushi_articles = pd.read_csv("../../process_articles/process_qiushi_articles/validated_articles.csv")
asx_articles = pd.read_csv("../../process_articles/process_asx_articles/validated_asx_articles.csv")
```

```python
def plot_publication_volume(df, source_name):
    """Plot the number of articles published in a quarter over time"""

    articles_per_quarter = get_articles_per_quarter(df)

    fig, ax = plt.subplots()

    xvals = articles_per_quarter.keys()
    yvals = articles_per_quarter.values()

    ax.plot(xvals, yvals)
    ax.set_title(f"{source_name} Articles Per Quarter")
    ax.set_xlabel('Quarter')
    ax.set_ylabel("Articles")

    ax.tick_params(axis='x', rotation=50)
    ax.grid()

    plt.savefig(f"/Users/calebharding/Documents/BYU/2023-2024/China_Project/reports/visuals/publication_volume/{source_name}_publication_volume.png", bbox_inches='tight')
    plt.show()
    
```

```python
articles_per_quarter = get_articles_per_quarter(qiushi_articles)
article_counts = pd.Series(articles_per_quarter.values())
article_counts.describe()
```

```python
articles_per_quarter = get_articles_per_quarter(asx_articles)
article_counts = pd.Series(articles_per_quarter.values())
article_counts.describe()
```

```python
plot_publication_volume(qiushi_articles, "Qiushi")
```
