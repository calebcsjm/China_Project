### Load data, convert format to match other files

```python
econ_data = pd.read_csv("Quarterly.csv", index_col=0)
```

```python
def date_flipper(quarter_year):
    date = quarter_year.split(" ")

    # the first column is just "Indicators," so leave that one be
    if len(date) == 1:
        return quarter_year
    
    year_quarter = date[1] + " " + date[0][1] + date[0][0]
    return year_quarter

def percent_gdp_growth(row):
    return 1

# flip dates to match my normal style
econ_data.rename(mapper=date_flipper, axis="columns", inplace=True)

# transpose the df and edit to match the format of the other calculations, with year_quarter as a column
econ_data = econ_data.transpose()
econ_data = econ_data.reset_index()
econ_data.rename(columns={"index":"year_quarter"}, inplace=True)
econ_data = econ_data.rename_axis(None, axis=1)
```

### Calculate additional values

```python
# lag the gdp so I can calculate percent growth
econ_data['L1 gdp current quarter'] = econ_data["Gross Domestic Product Current Quarter(100 million yuan)"].shift(-1)
# calculate percent growth
econ_data["quarter_gdp_growth"] = (econ_data["Gross Domestic Product Current Quarter(100 million yuan)"] - econ_data['L1 gdp current quarter']) / econ_data['L1 gdp current quarter'] * 100
```
